```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
import pandas as pd
import numpy as np
```

```python
URL = "https://www.sogo.com.hk/tc/catalog/mother-child/infant-feeding/infant-formula/illuma"
html = requests.get(URL)
```

```python
URL_source_code=html.text
```

```python
soup = BeautifulSoup(URL_source_code, "html.parser")
soup
```

```python
soup.find("div",class_="product-container").find("a",class_="content").find("div", class_="title").text.strip()
```

```python
soup.find("div", class_="product-price").find("div", class_="regular").text.strip()
```

```python
product_content = soup.find_all("div",class_="product-container")
product_content
```

```python
data_list = []
for i in product_content: #pay attention to the find_all
    title = i.find("a",class_="content").find("div", class_="title").text.strip()
    reg_price = i.find("a",class_="content").find("div", class_="product-price").find("div", class_="regular").text.strip()
    
    if i.find("div", class_="product-price").find("div", class_="discount"):
        dis_price = i.find("div", class_="product-price").find("div", class_="discount").text.strip()
    else:
        dis_price = None
    
    data_results= {
        "Title": title,
        "Reg Price": reg_price,
        "Discount": dis_price,
        "Supplier": "Sogo"
    }
    data_list.append(data_results)
data_list
```

```python
df = pd.DataFrame(data_list)
df
```

```python
df['Price']= np.where(df["Discount"].isnull(),df["Reg Price"],df["Discount"])
df['Price'] = df['Price'].replace({'\$':''},regex=True).str.replace(',', '').astype(float)
df
```

```python
df["Title"] = df["Title"].str.lower() 
```

```python
df['Series']=np.where(df.Title.str.contains('luxa'),"Luxa",
                np.where(df.Title.str.contains('hmo'),"Luxa",
                np.where(df.Title.str.contains('organic'),'Organic',
                np.where(df.Title.str.contains('atwo'),'Atwo',
                np.where(df.Title.str.contains('ha'),'HA','Others')))))
df
```

```python
df['Can'] = 6
df
```

```python
df['Stage']=np.where(df.Title.str.contains('stage 1'),"1",
                np.where(df.Title.str.contains('stage 2'),"2",
                np.where(df.Title.str.contains('stage 3'),'3',
                np.where(df.Title.str.contains('stage 4'),'4',         
                np.where(df.Title.str.contains('ha'),'1','Others')))))
df
```

```python
df['Parellel Import'] = False
df
```

```python
df['Unit Price']=df['Price']/df['Can']
```

```python
df['Merchant'] = "Chain"
df
```

```python
df.drop(["Reg Price","Discount"], axis=1, inplace = True)
```

```python
df
```

```python
df.to_csv (r'sogo.csv', index = False, header=True)
```
