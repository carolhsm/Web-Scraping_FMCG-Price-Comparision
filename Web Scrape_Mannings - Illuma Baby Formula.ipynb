```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
import pandas as pd
import numpy as np
```

```python
URL = "https://www.mannings.com.hk/brands/illuma/c/ILLUMA"

driver = webdriver.Chrome('./chromedriver')
```

```python
driver.get(URL) #replacing the requests.get step
subhtml = driver.page_source #replacing the html.text
```

```python
soup = BeautifulSoup(subhtml, "html.parser")
soup
```

```python
soup.find("h2", class_="prod_description").find("a", class_="pull-left product_name").text.strip()
```

```python
soup.find("div", class_="pdp_price_section pull-right text-right").find("p", class_="price").text.strip()
```

```python
product_content = soup.find_all("div",class_="product_content")
```

```python
product_title = soup.find("h2", class_="prod_description").find("a", class_="pull-left product_name").text.strip()
```

```python
product_price = soup.find("div", class_="details").find("div",class_="product_price").find("div", class_="pdp_price_section pull-right text-right").find("p", class_="price").text.strip()
```

```python
data_list = []
for i in product_content: #pay attention to the find_all
    title = i.find("h2", class_="prod_description").find("a", class_="pull-left product_name").text.strip()
    price = i.find("div", class_="details").find("div",class_="product_price").find("div", class_="pdp_price_section pull-right text-right").find("p", class_="price").text.strip()
    data_results= {
        "Title": title,
        "Price": price,
        "Supplier": "Mannings"
    }
    data_list.append(data_results)
data_list
```

```python
df = pd.DataFrame(data_list)
df
```

```python
df["Title"] = df["Title"].str.lower() 
```

```python
df['Brand']=np.where(df.Title.str.contains('illuma'),"Illuma","Others")
df
```

```python
df['Series']=np.where(df.Title.str.contains('luxa'),"Luxa",
                np.where(df.Title.str.contains('hmo'),"Luxa",
                np.where(df.Title.str.contains('organic'),'Organic',
                np.where(df.Title.str.contains('atwo'),'Atwo',
                np.where(df.Title.str.contains('ha'),'HA','Others')))))
df
```

```python
df['Stage']=np.where(df.Title.str.contains('stage 1'),"1",
                np.where(df.Title.str.contains('stage 2'),"2",
                np.where(df.Title.str.contains('stage 3'),'3',
                np.where(df.Title.str.contains('stage 4'),'4',         
                np.where(df.Title.str.contains('ha'),'1','Others')))))
df
```

```python
df['Gram']=df.Title.str[-4:]
df
```

```python
df['Unit'] = 1
df
```

```python
df['Parellel Import'] = False
df
```
